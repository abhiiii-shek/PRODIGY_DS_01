```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("C:/Users/ABHISHEK YADAV/Desktop/population.csv")
```

```python
#List of contries to plot on X axis
countries_to_plot = ['Canada', 'Germany', 'United States' , 'Japan', 'Brazil', 'France']

#Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_to_plot)]

#Creating the bar chart
plt.bar (filtered_data['Country Name'], filtered_data['2021'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2021')
plt.title('Population of selected countries in 2021')

#Rotating the x-axis labels to avoid overlap
plt.xticks(rotation=35, ha='right')
plt.show()
```
